##  Loading all dependencies
Here we load all the libraries we will need for our first regression problem.

```python
import numpy as np
import matplotlib.pyplot as plt

import torch
from torchmetrics.functional import r2_score, mean_squared_error, mean_absolute_error
from torch import nn
from torch.utils.data import Dataset, DataLoader, random_split
from collections import namedtuple
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
# check if a GPU is available
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
DEVICE
```

# 1  Linear regression 

## 1.1  Linear model 

### 1.1.1 Data generation
Here we generate some artificial data using a linear function and adding some noise to it.


```python
def f_linear(x, m, b):
    return m * x + b
```

```python
# number of points
npoints = 500

# noise parameter
dmax = 1

# interval
min_val = 0
max_val = 9

# slope and offset
m = 2.0
b = 5.00
```

```python
# generate data 
x = np.linspace(min_val, max_val, npoints)
y = f_linear(x, m, b) + np.random.randn(npoints) * dmax
y_orig = f_linear(x, m, b)

# make the data two-dimensional (expected by the model)
x = x.reshape(-1,1)
y = y.reshape(-1,1)
y_orig = y_orig.reshape(-1,1)
```

Let's plot the data and have a look!

```python
plt.plot(x,y, 'x', label = 'data')
plt.plot(x,y_orig,  label = 'linear',c="red")
plt.legend(fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### 1.1.2  Datasets and DataLoaders


```python
class Data(Dataset):
  """Dataset wrapper. Takes numpy arrays and stores them as torch tensors.
    Data is transformed to the device."""
  def __init__(self, x: np.ndarray,
                y: np.ndarray,
                device: str = DEVICE):
    self.x = torch.from_numpy(x).float().to(device)
    self.y =  torch.from_numpy(y).float().to(device)
    self.len = self.x.shape[0]
  
  def __getitem__(self, index: int) -> tuple:
    return self.x[index].float(), self.y[index].float()
  
  def __len__(self) -> int:
    return self.len
```

Now let's divide the full dataset into two smaller datasets for training and validation

```python
# define fraction of data used for training
train_fraction = .8
n_train = int(train_fraction * npoints)

# get indices for training and test set
ids = np.arange(npoints)
np.random.shuffle(ids)
ids_train, ids_test  = np.split(ids, [n_train])

# generate datasets for training and testing        
train_data = Data(x[ids_train], y[ids_train])
test_data = Data(x[ids_test], y[ids_test])
```

```python
print("The length of train data is:",len(train_data))
print("The length of test data is:",len(test_data))
```

```python
batch_size = 16
trainloader = DataLoader(train_data, 
                         batch_size=batch_size, 
                         shuffle=True)
testloader = DataLoader(test_data, 
                         batch_size=batch_size, 
                         shuffle=True)
```

Let's plot training and validation data!

```python
plt.plot(test_data.x,test_data.y, 'x', label = 'test')
plt.plot(train_data.x,train_data.y, 'x', label = 'train')
plt.plot(x, y_orig,  label = 'linear',c="red")
plt.legend(fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### 1.1.3  Building and training a model

```python
# fit the data with a linear regression model:
# y(w,x) = b + w1*x1 + w2*x2 + ... + wp*xp
# here we only fit b and w1, since we have 1D data
class LinearRegression(torch.nn.Module):
    def __init__(self):
        super(LinearRegression, self).__init__()
        self.linear = torch.nn.Linear(in_features=1,
                                      out_features=1,bias=True)

    def forward(self, x):
        out = self.linear(x)
        return out
```

```python
learningRate = 0.01 
n_epochs = 100

# generate a model and move it to the device
model = LinearRegression().to(DEVICE)
```

We can print the model to get some information.

```python
print(model)
```

Let's see how the untrained model performs!

```python
# Get model predictions for the test set
# we disable gradients because we are not in training mode here
# and get problems with numpy functions (used in the plotting routines)
with torch.no_grad():
    y_pred_test = model(test_data.x)

plt.plot(train_data.x ,train_data.y, 'x', label = 'train')
plt.plot(test_data.x, test_data.y, 'x', label = 'test')

# Plot the predicted data 
plt.plot(test_data.x, y_pred_test,c="blue", label = 'predictions',linewidth=2)
plt.plot(x,y_orig,  label = 'linear',c="red",linestyle="dashed",linewidth=2)
plt.legend(fontsize=15)

plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
print('R2 score = %8.5f' % r2_score(y_pred_test, test_data.y))
```

```python
# criterion to computes the loss between input and target
criterion = nn.MSELoss()

# optimizer that will be used to update weights and biases
optimizer = torch.optim.SGD(model.parameters(), lr=1e-2)
```

This is how a standard training loop in pytorch looks like.

```python
n_epochs = 1000
losses = np.empty(n_epochs)
n_log = 50

for epoch in range(n_epochs):
  running_loss = 0.0
  for i, (inputs, labels) in enumerate(trainloader):
    outputs = model(inputs)
    
    # compute loss
    loss = criterion(outputs, labels)   
    
    # set optimizer to zero grad to remove previous gradients
    optimizer.zero_grad()    # backward propagation
    
    # get gradients
    loss.backward()

    # gradient descent
    optimizer.step() 
    
    losses[epoch] = loss.item() # store loss

  if epoch % n_log == 0:
    print(f'Epochs:{epoch :5d} | ' \
          f'Loss: {loss :.10f}')
```

```python
plt.plot(losses)
plt.show()
```

### 1.1.4  Model evaluation


```python
# get predictions for training and test set
with torch.no_grad():
    y_pred_train = model(train_data.x)
    y_pred_test = model(test_data.x)
```

```python
plt.plot(train_data.x ,train_data.y, 'x', label = 'train')
plt.plot(test_data.x, test_data.y, 'x', label = 'test')

# Plot the predicted data 
plt.plot(test_data.x, y_pred_test,c="blue", label = 'predictions',linewidth=2)
plt.plot(x,y_orig,  label = 'linear',c="red",linestyle="dashed",linewidth=2)
plt.legend(fontsize=15)

plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
mse_test = mean_squared_error(test_data.y,y_pred_test)
mse_train = mean_squared_error(train_data.y,y_pred_train)

mae_test = mean_absolute_error(test_data.y,y_pred_test)
mae_train = mean_absolute_error(train_data.y,y_pred_train)

print('MSE test  = %8.4f    MAE test  = %8.4f' % (mse_test,mae_test))
print('MSE train = %8.4f    MAE train = %8.4f' % (mse_train,mae_train))
```

```python
print('R2 score = %8.5f' % r2_score(y_pred_test, test_data.y))
```

Let's see what the parameters the model has learned. 

```python
for layer in model.children():
    if isinstance(layer, nn.Linear):
        print("Learned m:", layer.state_dict()['weight'])
        print("Learned b:", layer.state_dict()['bias'])
print("Exact m:", m)
print("Exact b:",b)
```

## 1.2 Experiments

### 1.2.1 Different datasets

```python
RawData = namedtuple("RawData", ["x", "y", "y_orig"])
```

```python
# Generate new data
def generate_data(n_points=500, d_max=1, min_val=0, max_val=9, m=2.0, b=5.00):
    x = np.linspace(min_val, max_val, n_points)
    y = f_linear(x, m, b) + np.random.randn(npoints) * d_max
    y_orig = f_linear(x, m, b)

    x = x.reshape(-1,1)
    y = y.reshape(-1,1)
    y_orig = y_orig.reshape(-1,1)
    return RawData(x, y, y_orig)
```

```python
d_max_data = {
    "1" : generate_data(d_max=1),
    "10": generate_data(d_max=10),
    "20": generate_data(d_max=20)
}
```

```python
def plot_data(data, key="Data", save=False, name=None):
    plt.figure(figsize=(15, 3))
    for (i, (key, data)) in enumerate(data.items()):
        plt.subplot(1, 3, i+1)
        plt.title(f"d_max = {key}")
        plt.plot(data.x, data.y, 'x', label = 'Perturbed')
        plt.plot(data.x, data.y_orig,  label = 'Original',c="red")
        plt.legend(fontsize=15)
        plt.xlabel('x', fontsize=15)
        plt.ylabel('y', fontsize=15)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)

    if save:
        if name != None:
            plt.savefig(name)
        else:
            print("Saving failed! Please give a name.")
    
    plt.tight_layout()
    plt.show()
```

```python
# Plot the data to see if it was generated correctly
plot_data(d_max_data, save=True, name="dmax_dataset.png")
```

```python
Datasets = namedtuple("Datasets", ["train", "test"])
Dataloaders = namedtuple("Dataloaders", ["train", "test"])
```

```python
# Make Datasets and DataLoaders with the new data
def create_dataset(data, train_fraction=0.8):
    npoints = len(data.x)
    n_train = int(train_fraction * npoints)

    ids = np.arange(npoints)
    np.random.shuffle(ids)
    ids_train, ids_test  = np.split(ids, [n_train])

    train_data = Data(data.x[ids_train], data.y[ids_train])
    test_data = Data(data.x[ids_test], data.y[ids_test])

    return Datasets(train_data, test_data)

def create_dataloader(datasets, batch_size=16):
    train_loader = DataLoader(datasets.train, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(datasets.test, batch_size=batch_size, shuffle=True)
    return Dataloaders(train_loader, test_loader)
```

```python
d_max_datasets = {key: create_dataset(data) for (key, data) in d_max_data.items()}
d_max_dataloaders = {key: create_dataloader(datasets) for (key, datasets) in d_max_datasets.items()} 
```

```python
# Define a linear model and train
def create_and_train(train_loader, lr=0.01, nepochs=1000):
    model = LinearRegression().to(DEVICE)
    criterion = nn.MSELoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=lr)
    losses = np.empty(nepochs)

    for epoch in range(nepochs):
        running_loss = 0.0
        for i, (inputs, labels) in enumerate(train_loader):
            outputs = model(inputs)
            loss = criterion(outputs, labels)   
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        losses[epoch] = running_loss / len(train_loader)
    
        if epoch % 100 == 0:
            print(f'Epochs:{epoch :5d} | Loss per iteration: {losses[epoch] :.10f}')

    print("Finished!\n")
    return model
```

```python
trained_models = {key: create_and_train(dataloader.train) for (key, dataloader) in d_max_dataloaders.items()}
```

```python
# make predictions and evaluate
# get predictions for training and test set

def plot_predictions(models, data, datasets, save=False, name=None):

    plt.figure(figsize=(12,7))

    for (i, ((key, dataset), model)) in enumerate(zip(datasets.items(), models.values())):
        train_dataset, test_dataset = dataset
        with torch.no_grad():
            y_pred_train = model(train_dataset.x)
            y_pred_test = model(test_dataset.x)

        plt.subplot(2, 3, i+1)
        plt.title(f"dmax = {key}")
        plt.plot(train_dataset.x[::4] ,train_dataset.y[::4], 'x', label = 'train')
        plt.plot(train_dataset.x, y_pred_train, c="black", label = 'predictions',linewidth=2)
        plt.plot(data[key][0], data[key][2], c="red", label = 'linear', linestyle="dashed",linewidth=2)
        plt.legend(fontsize=15)

        plt.xlabel('x', fontsize=15)
        plt.ylabel('y', fontsize=15)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)

        plt.subplot(2, 3, i+4)
        plt.title(f"dmax = {key}")
        plt.plot(test_dataset.x, test_dataset.y, 'x', label = 'test')
        plt.plot(test_dataset.x, y_pred_test, c="black", label = 'predictions',linewidth=2)
        plt.plot(data[key][0], data[key][2], c="red", label = 'linear', linestyle="dashed",linewidth=2)
        plt.legend(fontsize=15)

        plt.xlabel('x', fontsize=15)
        plt.ylabel('y', fontsize=15)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)

    if save:
        if name != None:
            plt.savefig(name)
        else:
            print("Saving failed! Please give a name.")


    plt.tight_layout()
    plt.show()


def plot_prediction(model, data, dataset, key="Model", save=False, name=None):

    plt.figure(figsize=(8, 4))
    train_dataset, test_dataset = dataset
    with torch.no_grad():
        y_pred_train = model(train_dataset.x)
        y_pred_test = model(test_dataset.x)

    plt.subplot(1, 2, 1)
    plt.title(f"{key} (Train)")
    plt.plot(train_dataset.x , train_dataset.y, 'x', label = 'train')
    plt.plot(train_dataset.x, y_pred_train, c="black", label = 'predictions',linewidth=2)
    plt.plot(data.x, data.y_orig, c="red", label = 'linear', linestyle="dashed",linewidth=2)
    plt.legend(fontsize=15)

    plt.xlabel('x', fontsize=15)
    plt.ylabel('y', fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.subplot(1, 2, 2)
    plt.title(f"{key} (Test)")
    plt.plot(test_dataset.x, test_dataset.y, 'x', label = 'test')
    plt.plot(test_dataset.x, y_pred_test, c="black", label = 'predictions',linewidth=2)
    plt.plot(data.x, data.y_orig, c="red", label = 'linear', linestyle="dashed",linewidth=2)
    plt.legend(fontsize=15)

    plt.xlabel('x', fontsize=15)
    plt.ylabel('y', fontsize=15)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    if save:
        if name != None:
            plt.savefig(name)
        else:
            print("Saving failed! Please give a name.")


    plt.tight_layout()
    plt.show()
```

```python
plot_predictions(trained_models, d_max_data, d_max_datasets, save=True, name="dmax_predictions.png")
```

```python
## Check the learned parameters
for (key, model) in trained_models.items():
    for layer in model.children():
        if isinstance(layer, nn.Linear):
            print(f"d_max = {key}")
            print("Learned m:", layer.state_dict()['weight'].item())
            print("Learned b:", layer.state_dict()['bias'].item(), end="\n\n")
```

```python
def print_evaluation(model, dataset, label="Model"):
    train_dataset, test_dataset = dataset
    with torch.no_grad():
        y_pred_train = model(train_dataset.x)
        y_pred_test = model(test_dataset.x)

    mse_test = mean_squared_error(test_dataset.y, y_pred_test)
    mse_train = mean_squared_error(train_dataset.y, y_pred_train)

    mae_test = mean_absolute_error(test_dataset.y, y_pred_test)
    mae_train = mean_absolute_error(train_dataset.y, y_pred_train)

    print(f"Dataset (dmax={label})")
    print('R2 score = %8.5f' % r2_score(y_pred_test, test_dataset.y))
    print('MSE train = %8.4f    MAE train = %8.4f' % (mse_train,mae_train))
    print('MSE test  = %8.4f    MAE test  = %8.4f' % (mse_test,mae_test), end="\n\n")
    
def print_evaluations(models, datasets):
    for ((key, dataset), model) in zip(datasets.items(), models.values()):
        print_evaluation(model, dataset, label=key)
```

```python
print_evaluations(trained_models, d_max_datasets)
```

```python
class PTMinMaxScaler(object):
    """"Scaler class for min-max scaling."""
    def __init__(self, tensor, new_max, new_min):
        self.v_min, self.v_max = tensor.min(), tensor.max()
        self.new_max, self.new_min = new_max, new_min

    def scale(self, tensor):
        return (tensor - self.v_min)/(self.v_max - self.v_min)*(self.new_max - self.new_min) + self.new_min

    def inverse_scale(self, tensor):
        return (tensor - self.new_min) * (self.v_max - self.v_min)/(self.new_max - self.new_min) + self.v_min
```

```python
# example usage of scaler class
x = np.linspace( -1, 2, 5)

# fit the scaler
x_scaler = PTMinMaxScaler(x, 1. , 0.)
x_scaled = x_scaler.scale(x)

print(x,"\n", x_scaled, "\n", x_scaler.inverse_scale(x_scaled))
```

```python
# scale the data and divide into training and test data
scaled_data_scaler = PTMinMaxScaler(scaled_data.x, 1.0, 0.0)
scaled_data_with_scaler = RawData(scaled_data_scaler.scale(scaled_data.x), scaled_data_scaler.scale(scaled_data.y), scaled_data_scaler.scale(scaled_data.y_orig))
```
